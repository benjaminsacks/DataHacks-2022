```python
import numpy as np
import pandas as pd
from collections import defaultdict
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("advanced_trainset.csv")
df
```

```python
df["Sentiment"].value_counts()
```

```python
cleaned = df.copy(deep=True)
cleaned["Sentence_cleaned"] = cleaned["Sentence"].str.lower()
# cleaned["Sentence_cleaned"] = cleaned["Sentence_cleaned"].str.replace(r",|\.", " ")
cleaned
```

```python
words = defaultdict(int)
for i in cleaned["Sentence_cleaned"]:
    for i in i.split():
        words[i] += 1

words
```

```python
len(words)
```

```python
word_index = {}
counter = 0
for i in words:
    if words[i] > 1:
        word_index[i] = counter
        counter += 1
word_index[" "] = counter
```

```python
len(word_index)
```

```python
word_count = []
for sentence in cleaned["Sentence_cleaned"]:
    result = [0] * 5225
    for word in sentence.split(" "):
        if word in word_index:
            result[word_index[word]] += 1
    word_count.append(result)
```

```python
df = pd.DataFrame(word_count)
df["length"] = cleaned["Sentence_cleaned"].str.split().apply(len)
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df,cleaned["Sentiment"],test_size=0.75,random_state=69)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
clf = DecisionTreeClassifier(random_state=0)
clf.fit(X_train, y_train)
accuracy_score(clf.predict(X_test), y_test)
```

```python
clf = GaussianProcessClassifier()
clf.fit(X_train, y_train)
accuracy_score(clf.predict(X_test), y_test)
```

```python
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
accuracy_score(clf.predict(X_test), y_test)
```

```python
clf = AdaBoostClassifier()
clf.fit(X_train, y_train)
accuracy_score(clf.predict(X_test), y_test)
```

```python
clf = GaussianNB()
clf.fit(X_train, y_train)
accuracy_score(clf.predict(X_test), y_test)
```
